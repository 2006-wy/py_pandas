```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Matplotlib

```python
#创建画布
plt.figure(figsize=(8, 6))#设置画布大小 8英寸宽，6英寸高

plt.plot([1, 2, 3], [4, 5, 6])#绘制简单的线图 x轴数据为[1, 2, 3]，y轴数据为[4, 5, 6]

```

```python
#折线图
x = range(60)#生成0到59的整数序列作为x轴数据
y_bijing = [np.random.uniform(20, 30) for i in x]#生成60个随机数作为北京的温度数据，范围在20到30之间 uniform()函数用于生成指定范围内的随机浮点数

plt.figure(figsize=(10, 6), dpi=100)#设置画布大小 10英寸宽，6英寸高 dpi参数设置图像的分辨率为100像素每英寸
plt.plot(x, y_bijing, label='Beijing', color='blue')#绘制北京的折线图，x轴数据为x，y轴数据为y_bijing，label参数用于设置图例标签 color参数用于设置线条颜色

#添加标题和标签
x_ticks_labels = [f'{i}h' for i in x]#生成x轴刻度标签，格式为0h, 1h, ..., 59h
y_ticks = range(60)#生成y轴刻度，范围从0到59

plt.yticks(y_ticks[::10])#设置y轴刻度
plt.xticks(x[::5], x_ticks_labels[::5])#设置x轴刻度和标签，每隔5个刻度显示一次 参数x[::5]表示从x中每隔5个元素取一个，x_ticks_labels[::5]表示从x_ticks_labels中每隔5个元素取一个
#添加网格
plt.grid(True, linestyle='--', alpha=0.5)#显示网格线 linestyle参数设置网格线的样式为虚线 alpha参数设置网格线的透明度为0.5


#添加标题和标签
plt.title('Temperature in Beijing')#设置图表标题
plt.xlabel('Time (hours)')#设置x轴标签
plt.ylabel('Temperature (°C)')#设置y轴标签

#图像保存
plt.savefig('beijing_temperature.png')#将当前图像保存为beijing_temperature.png

#添加图例
plt.legend(loc =10)#显示图例  作用与plt.show的区别在于plt.legend()用于显示图例，而plt.show()用于显示整个图像。plt.legend()需要在绘制图形后调用，以便显示图例，而plt.show()通常在所有绘图命令之后调用，以显示最终的图像。
plt.show()#显示图像
```
