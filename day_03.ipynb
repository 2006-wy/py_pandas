```python
#导包
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 1.Pandas 基础操作

```python
#调用数据
data = pd.read_csv('C:/Users/35986/Desktop/stock_day.csv')
print(pd.DataFrame(data))
#数据清洗
print('----------------------------')
data.drop(columns=['ma5', 'ma10','ma20','v_ma5','v_ma10','v_ma20'],axis=0, inplace=True)#删除不需要的列
print(data)
```

### 1.2 索引操作

```python
#1:Find data by rows and columns
print(data['open'])
print('----------------------------')
print(data['open']['2018-01-02'])
```

```python
# 2:loc() 行索引 和 列名
data.loc['2018-02-27','high']
print('----------------------------')
data.loc['2018-02-27':'2018-02-20','open':'high']
```

```python
#3:iloc() 行号 和 列索引  先行后列
print(data.iloc[0,1])

print('----------------------------')
print(data.iloc[0:5,1:4])
```

```python
#赋值
data.loc['2018-02-27','high'] = 100
print(data.loc['2018-02-27','high'])
```

```python
#基于open 做升序
print(data.sort_values(by='open',ascending=True))
```

```python
data.sort_values(by=['open','high'],ascending=[True,True])
```

```python
data.sort_index(ascending=True)#基于行索引升序
```

```python
print(data.open.sort_values(ascending=True))#基于open列升序
print('----------------------------')
print(data.open.sort_index(ascending=True))
```

## 2.算法运算


```python
print(data.close)
print('----------------------------')
print(data.close + 2)
```

```python
#筛选
print(data.open[data.open > 23])
print('----------------------------')
print(data[(data.open > 23) & (data.close < 24)])
```

```python
#固定值筛选open等于23.53 或者 close等于24.67
print(data[(data.open == 23.53) | (data.close == 24.67)])
```

```python
#isin()方法筛选open等于23.53 或者 close等于24.67
print(data[data.open.isin([23.53]) | data.close.isin([24.67])])
```

###  3.统计运算

```python
print(data.sum())#求和
print('----------------------------')
print(data.mean())#求平均值
print('----------------------------')
print(data.max())#求最大值
print('----------------------------')
print(data.high.sum())#求high列的和
```

```python
#apply()方法对数据进行批量处理
def add_10(x):
    return x.max() - x.min() 
print(data[['open','close']])
print('----------------------------')
print(data[['open','close']].apply(add_10))
```
