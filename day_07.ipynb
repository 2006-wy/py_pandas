```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('uniqlo.csv')
```

# 数据分组

```python
#agg  -> aggregate聚合
#格式 df.groupby(['分组1', '分组2'...]).agg({'列1': '函数1', '列2': '函数2'...})
```

```python
#场景1：按照单列分组
df.groupby(['city'])
```

```python
#场景2：按照多列分组
df.groupby(['city', 'gender'])
```

```python
#如何查看分组后的数据
#分组后的数据是一个DataFrameGroupBy对象，可以使用get_group方法查看某个分组的数据
#查看city为Beijing的分组数据
df.groupby(['city','channel']).get_group(('北京', '线下'))   
```

```python
#场景4：分组后进行聚合
#按照city和销售渠道分组，计算每个分组的销售额
df.groupby(['city', 'channel']).agg({'revenue': 'sum'})
```

```python
#分组过滤
#销售额平均大于200的分组

df.groupby('city').filter(lambda x: x.revenue.mean() > 200)
```
