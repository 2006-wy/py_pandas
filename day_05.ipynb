```python
import pandas as pd
df = pd.read_csv('E:\\35986\\PycharmProjects\\PythonProject\\Pandas\\1960-2019全球GDP数据.csv',encoding = 'gbk')
df1 = df.head(10)
```

# DataFrame 数据的增删改查

```python
# 5.1 
print(df1)
```

```python
#拷贝数据
df_copy = df1[0:5].copy()
print(df_copy)
```

```python
#新增列
#1.直接赋值
df_copy['c1'] = 1
#2.使用assign方法增加n列
df_copy = df_copy.assign(
    c2=2,
    c3 = ['cxk','cxk','cxk','cxk','cxk'],
    c4 = lambda x: x['c1'] + x['c2']
                         )
print(df_copy)
```

```python
#删除列
#del方法 与 drop方法
#del方法只能删除一列，且无法返回删除后的数据 且 永远修改原数据
#drop方法可以删除多列，且可以返回删除后的数据，且可以选择是否修改原数据
```

```python
df4 = df_copy.copy()
print(df4)
print('------------------')
#使用del方法删除c1列
#del df4['c1']
#print(df4)
print('------------------')
#使用drop方法删除c2列
df5 = df4.drop(index=[0],axis=1,inplace=False)
print(df5)
df6 = df4.drop(columns=['c2'],axis=1,inplace=False)
print(df6)
```

```python
#去重
#添加重复数据
df7 = pd.concat([df_copy,df_copy],ignore_index=True)#ignore_index参数：是否重新索引 concat:连接
print(df7)
print('------------------')
#使用drop_duplicates方法去重
df8 = df7.drop_duplicates(inplace=True)#设置inplace=True会修改原数据
print(df8)
```

```python
#改
df7['c1'] = 'hahahah'#直接修改原数据
print(df7)
print('------------------')
#使用replace方法修改数据
df9 = df7.replace('hahahah','yuno',inplace=False)#设置inplace=True会修改原数据 'hahahah替换为yuno'
print(df9)
```

```python
#查
#使用loc方法根据标签查询数据
print(df9.loc[0])#查询第一行数据
print(df9.loc[0:2])#查询第一行到第三行数据
print(df9.loc[0,'c3'])#查询第一行c3列数据
print('------------------')
#query方法根据条件查询数据
print(df9.query('c2 == 2'))#查询c2列等于2的数据
print(df9.query('c2 > 1'))#查询c2列大于1的数据
```
