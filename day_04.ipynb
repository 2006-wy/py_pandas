### 5.1Pandas操作csv文件

```python
#读取数据
df = pd.read_csv('stock_day.csv',sep=',',usecols=['open','high','close','low'])
#查看数据
print(df)
```

```python
#写入数据
df[:10].to_csv('./stock_day_1.csv',index=False,sep=',')
print('数据写入成功！')

df.to_csv('stock_day_2.csv',index=True,sep=',')
```

```python
#tsv文件和csv文件的区别
#csv文件以逗号分隔数据，而tsv文件以制表符Tab分隔数据。csv文件适用于大多数数据处理工具，而tsv文件在处理包含逗号的数据时更为方便。
```

```python
#特殊的csv文件->tsv文件
df[:5].to_csv('stock_day_3.tsv',index=True,sep='\t')
print('tsv文件写入成功！')

df2 = pd.read_csv('stock_day_3.tsv',sep='\t',index_col=0)
print(df2)
print(type(df2))
```
