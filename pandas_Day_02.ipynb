```python
#导包
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Pandas数据类型

```python
#1.创建DataFrame对象
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
})
print(data)
```

```python
data.info()
print('------------------------')
print(data.describe())
```

```python
#datetime数据类型
df = pd.DataFrame(['2024-01-01', '2024-02-01', '2024-03-01'], dtype='datetime64[ns]', columns=['Date'])
print(df)
print(df.dtypes)
```

```python
#timedelta数据类型时间差
start_date = pd.to_datetime('2024-01-01')
end_date = pd.to_datetime('2024-01-10')

print(end_date - start_date)      
#输出时间差   能当数字使用吗？

print(type(end_date - start_date))
```

```python
#category数据类型
s1 = pd.Series(['A', 'B', 'A', 'C', 'B'], dtype='category')
print(s1)
print(s1.dtypes)
print(type(s1))
```
