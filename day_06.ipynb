# 缺失值处理

```python
#缺失值NAN,NaN,nan
import numpy as np
import pandas as pd

data = pd.read_csv('./movie.csv')
print(data)
```

```python
#查介绍
print(data.columns)#查看列名
print('------------------')
print(data.info())#查看数据类型和缺失值情况

print('------------------')
print(data.describe())#查看数值型数据的统计信息
```

```python
#删除缺失值  dropna方法不会修改原数据       需要使用inplace参数
da1 = data.dropna()#删除所有含有缺失值的行
print(da1)
da2 = data.dropna(axis=1)#删除所有含有缺失值的列
print(da2)
print('------------------')
print(da1.info())#查看删除缺失值后的数据类型和缺失值情况
print('------------------')
print(da2.info())#查看删除缺失值后的数据类型和缺失值
```

```python
#判断是否存在缺失值
print(pd.isnull(data))#判断每个元素是否为缺失值，返回一个布尔型的DataFrame
print('------------------')
print(pd.notnull(data))#判断每个元素是否为非缺失值，返回一个布尔型的DataFrame
```

```python
#判断某列是否存在缺失值
print(np.all(pd.notnull(data)))#判断整个DataFrame是否存在缺失值，返回一个布尔值            ##########np.all()函数用于判断所有元素是否满足条件，这里判断是否所有元素都为非缺失值，如果存在缺失值则返回False  pd.notnull(data)返回一个布尔型的DataFrame，表示每个元素是否为非缺失值，np.all()函数对这个布尔型的DataFrame进行判断，如果所有元素都为True（即所有元素都是非缺失值），则返回True；如果存在至少一个元素为False（即存在缺失值），则返回False。
```

```python
#for循环的方法，用每列的平均值填充缺失值
for col in data.columns:
    if pd.api.types.is_number(data[col]) and  np.all(pd.notnull(data[col]))==False:#判断该列是否存在缺失值
        mean_value = data[col].mean()#计算该列的平均值
        data[col].fillna(mean_value,inplace=True)#使用平均值填充缺失值，inplace参数表示修改原数据
print(data)

print('------------------')
print(data.info())#查看填充缺失值后的数据类型和缺失值情况
```

```python
#? -> NaN -> 删
da3 = [
    ['a', 1, 2],
    ['b', 3, '?'],
    ['c', '?', 4],
    ['d', 5, 6]
]
df = pd.DataFrame(da3, columns=['A', 'B', 'C'])
print(df)
```

```python
#? -> NaN -> 删
df.replace('?', np.nan, inplace=True)#将?替换为NaN
df.dropna(inplace=True)#删除包含NaN的行
print(df)
```

```python
#数据合并
da4 = pd.concat([df, df], axis=0, ignore_index=True)#按行合并，ignore_index参数表示重新索引
print(da4)
print('------------------')
da5 = pd.concat([df, df], axis=1)#按列合并
print(da5)
```

```python
#满外连接->左表全集+右表全集+交集部分
da6 = pd.concat([df, df], axis=0, ignore_index=True, join='outer')#按行合并，join参数表示连接方式，outer表示满外连接
print(da6)
print('------------------')
#内连接->交集部分
da7 = pd.concat([df, df], axis=0, ignore_index=True, join='inner')#按行合并，join参数表示连接方式，inner表示内连接
print(da7)
```
