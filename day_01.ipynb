### 2.创建Numpy的ndarray对象

### 3.等比数列&等差数列

```python
#3.1 logspace(start(充当10的指数), stop（充当10的指数）, num=元素个数,base = 基数(默认为10))函数生成等比数列
arr8 = np.logspace(0, 3, num=4)
print(arr8)
```

```python
#3.2 linspace(start, stop, num=元素个数， endpoint=True(是否包含结束值),dtype=np.None )函数生成等差数列
arr9 = np.linspace(0, 10, num=5)
print(arr9)
print(arr9.dtype)
```

# 4.numpy内置函数

```python
# 4.1
arr11 = np.random.rand(3, 4) * 10
print(arr11)
```

```python
#ceil()函数向上取整
arr12 = np.ceil(arr11)
print(arr12)
```

```python
#floor()函数向下取整
arr13 = np.floor(arr11)
print(arr13)
print('-----------------------------')
#round()函数四舍五入
arr14 = np.round(arr11)
print(arr14)
print('-----------------------------')
#abs()函数取绝对值
arr15 = np.abs(arr11 - 5)
print(arr15)
print('-----------------------------')
#multiply(arr1,arr1)函数矩阵乘法(行列数要一致)
arr16 = np.multiply(arr11, arr11)
print(arr16)
print('-----------------------------')
print(arr11*arr11)#矩阵乘法(行列数要一致)同上
print('-----------------------------')
#where(判断，x,y)函数根据条件返回满足条件的元素的索引
arr17 = np.where(arr11 > 5,1,0)
print(arr17)
```

## 5.统计函数

```python
arr20 = np.arange(12).reshape(3, 4)
print(arr20)

#cumsum()函数计算数组元素的累积和
arr21 = np.cumsum(arr20)
print(arr21)

#sum()函数计算数组元素的和 axis=0按列求和，axis=1按行求和
arr22 = np.sum(arr20, axis=0)#按列求和
print(arr22)
arr23 = np.sum(arr20, axis=1)#按行求和
print(arr23)
```

```python
#去重函数unique()函数返回数组中唯一的元素
list1 = ([[1, 2, 3], [4, 5, 6], [3,9,8]])
print(list1)
arr24 = np.unique(list1)
print(arr24)
```

```python
#np.sort()函数对数组进行排序
arr25 = np.sort(list1)
print(arr25)
print('-----------------------------')
arr20.sort()
print(arr20)
```

```python
#乘积  行列数不一致
#@   or  dot()函数矩阵乘法
list2 = np.array([[1, 2], [3, 4], [5, 6]])
list3 = np.array([[1, 2, 3], [4, 5, 6]])
print(list2 @ list3)
print(list2.dot(list3))
print(np.dot(list2, list3))
```

# *DataFrame

```python
#字典+列表实现
info = {
    'name':['张三','李四','王五','赵六','钱七'],
    'age':[18,19,20,21,22],
    'city':['北京','上海','广州','深圳','杭州']
}

df1 = pd.DataFrame(info)
print(df1)
```

```python
#列表+元组实现
data = [('张三', 18, '北京'),
        ('李四', 19, '上海'),
        ('王五', 20, '广州'),
        ('赵六', 21, '深圳'),
        ('钱七', 22, '杭州')]
df2 = pd.DataFrame(data, columns=['name', 'age', 'city'])
print(df2)
```

###  -----简单案例-----

```python
#生成10名学生，5门功课的成绩（40-100之间的随机整数）
score_df = pd.DataFrame(np.random.randint(40,101,size = (10,5)),columns = ['语文','数学','英语','体育','音乐'])
index_names = ['Student'+str(i) for i in range(score_df.shape[0])]
score_df.index = index_names
print(score_df)
```

```python
# head()函数返回DataFrame的前n行，默认为5行
print(score_df.head())
print('-----------------------------')
print(score_df.head(3))#返回前3行
```

```python
# tail()函数返回DataFrame的后n行，默认为5行
print(score_df.tail())
print('-----------------------------')
print(score_df.tail(3))#返回后3行
```

```python
#info()函数查看DataFrame的基本信息
score_df.info()
print('-----------------------------')
#describe()函数查看DataFrame的统计信息
score_df.describe()
```
