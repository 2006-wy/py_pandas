# 交叉表和同学和透视表

```python
#交叉表
df = pd.DataFrame({'性别': ['男', '女', '男', '女', '男', '女'],
                     '是否吸烟': ['是', '否', '是', '否', '是', '否']})
crosstab = pd.crosstab(df['性别'], df['是否吸烟'])#交叉表
print(crosstab)
```

```python
df1 = pd.read_csv('uniqlo.csv')
#根据city和channel列创建交叉表计算销售总额
print(df1.groupby(['city', 'channel']).agg({'revenue': 'sum'}))
print('-----------------------------')
print(df1.groupby(['city', 'channel'], as_index=False).agg({'revenue': 'sum'}))
```

```python
#透视表pivot_table()
#根据city和channel列创建透视表计算销售总额
print(df1.pivot_table(index='city', columns='channel', values='revenue', aggfunc='sum'))
```

```python
house_df = pd.read_csv('LJdata.csv')
house_df.head()
```

```python
#修改列名
house_df.rename(columns={'价格': 'price', '标题': 'title'}, inplace=True)
house_df.head()
```

```python
#查看数据维度
print(house_df.shape)
```

```python
#找到价格最高和最低的房子
print(house_df.nlargest(1, 'price'))#nlargest()函数返回DataFrame中指定列的前n个最大值所在
print('-----------------------------')
print(house_df.nsmallest(1, 'price'))#nsmallest()函数返回DataFrame中指定列的前n个最小值所在
```
